## Data Visualization of IRIS Dataset

```python
# Importing necessary libraries...

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing IRIS dataset through pandas library..

data=pd.read_csv("IRIS.csv")
data.head()
```

```python
data.info()  # getting information about data
```

```python
data.shape  #getting shape of data
```

```python
data.isnull().sum() #to check whether any null value is present in data or not 
```

```python
data.describe() #to describe data
```

```python
data['species'].unique() #to get unique values of target_names
```

```python
column = data.columns[:-1].tolist()
label = ['SEPAL LENGTH', 'SEPAL WIDTH', 'PETAL LENGTH', 'PETAL WIDTH']
```

```python
data1=data.drop(['species'],axis=1)
```

### Lineplot for data and its features

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=data1,markers=True)
plt.title("Lineplot for Iris Dataset",fontsize=15)
```

```python
color=['red','green','blue','orange']
for i in range(len(label)):
    plt.figure(figsize=(10,5))
    sns.lineplot(data=data1.iloc[:,i],markers=True,color=color[i],marker='o')
    plt.title("Lineplot for {}".format(label[i]),fontsize=15)

```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=data.index,y=data1.iloc[:,-1],hue=data['species'],marker='o')
#plt.ylabel(label[3], fontsize=12)
plt.title("Lineplot for {}".format(label[-1]),fontsize=15)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=data1.iloc[:,0],y=data1.iloc[:,1],markers=True,color='green',marker='d')
plt.title("Lineplot {} vs {} ".format(label[0],label[1]),fontsize=15)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=data1.iloc[:,2],y=data1.iloc[:,3],markers=True,hue=data['species'],marker='d')
plt.title("Lineplot {} vs {} by category".format(label[2],label[3]),fontsize=15)
```

### Scatter plot for data and its features

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=data1,markers=True,s=50)
plt.title("Scatterplot for Iris Dataset",fontsize=15)
```

```python
m=['o','d','s']
for i in range(len(label)-1):
    plt.figure(figsize=(10,5))
    sns.scatterplot(data=data1.iloc[:,i],marker=m[i],color=color[i],alpha=0.7,s=90)
    plt.title("Scatterplot for {}".format(label[i]),fontsize=15)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data.index,data1.iloc[:,-1],hue=data['species'],marker='o',s=150,alpha=0.7)
plt.title("Scatterplot for Petal_width",fontsize=15)
```

### Regression plot for data and its features

```python
for i in range(len(label)):
    plt.figure(figsize=(10,5))
    sns.regplot(data.index,data1.iloc[:,i],color=color[i],marker='o')
    plt.title("Regression plot for {}".format(label[i]),fontsize=15)
```

### Distribution plot for data and its features

```python
plt.figure(figsize=(10,5))
sns.distplot(data1)
plt.title("Distribution plot for Iris Dataset ",fontsize=15)
```

```python
for i in range(len(label)):
    plt.figure(figsize=(10,5))
    sns.distplot(data1.iloc[:,i],color=color[i])
    plt.title("Distribution plot for {}".format(label[i]),fontsize=15)
```

### KDE plot for data and its features

```python
plt.figure(figsize=(10,5))
plt.title('KDEPLOT ', fontsize=15)
sns.kdeplot(data=data1,shade=True)
plt.show()
```

```python
for i in range(len(label)):
    plt.figure(figsize=(10,5))
    sns.kdeplot(data1.iloc[:,i],shade=True,color=color[i])
    plt.title("KDE plot for {}".format(label[i]),fontsize=15)
```

### Strip-plot for data and its features

```python
for i in range(len(color)):
    plt.figure(figsize=(10,5))
    plt.title('Stripplot for {}'.format(label[i]), fontsize=15)
    sns.stripplot(data.species, data[column[i]],s=12,alpha=0.5)
    plt.xlabel('SPECIES', fontsize=12)
    plt.ylabel(label[i], fontsize=12)
    plt.show()
```

### Boxplot for data and its features

```python
plt.figure(figsize=(10,5))
plt.title('Boxplot for IRIS data', fontsize=15)
sns.boxplot(data=data1)
plt.show()
```

```python
for i in range(len(color)):
    plt.figure(figsize=(10,5))
    plt.title('Boxplot for {}'.format(label[i]), fontsize=15)
    sns.boxplot(data[column[i]],color=color[i])
    plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Boxplot for {}'.format(label[3]), fontsize=15)
sns.boxplot(column[3], data=data1, palette='Set3', notch=True)
plt.show()
```

```python
for i in range(len(color)):
    plt.figure(figsize=(10,5))
    plt.title('Boxplot for {} by Category'.format(label[i]), fontsize=15)
    sns.boxplot(x=data['species'], y=column[i], data=data1, palette='Set2')
    plt.xlabel('SPECIES', fontsize=12)
    plt.ylabel(label[i], fontsize=12)
    plt.show()
```

### Violin plot for data and its features

```python
plt.figure(figsize=(10,5))
plt.title('Violinplot for IRIS data', fontsize=15)
sns.violinplot(data=data1)
plt.show()
```

```python
for i in range(len(color)):
    plt.figure(figsize=(10,5))
    plt.title('Violinplot for {} by Category'.format(label[i]), fontsize=15)
    sns.violinplot(data1.iloc[:,i], data=data1,color=color[i])
    plt.show()
```

```python
for i in range(len(label)):
    plt.figure(figsize=(10,5))
    plt.title('Violinplot for {}'.format(label[i]), fontsize=15)
    sns.violinplot(data['species'], column[i], data=data1, palette='Reds',inner='stick')
    plt.xlabel('SPECIES', fontsize=12)
    plt.ylabel(label[i], fontsize=12)
    plt.show()
```

### Swarmplot for data and its features

```python
plt.figure(figsize=(10,5))
plt.title('Swarmplot for IRIS Data', fontsize=15)
sns.swarmplot(data=data1, s=10, alpha=0.7)
plt.show()
```

```python
for i in range(len(label)):
    plt.figure(figsize=(10,5))
    plt.title('Swarmplot for {}'.format(label[i]), fontsize=15)
    sns.swarmplot(x=data[column[i]], s=12,color=color[i])
    plt.xlabel(label[i], fontsize=12)
    plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Swarmplot for {}'.format(label[2]), fontsize=15)
sns.swarmplot(x=data1[column[2]], y=data.species, s=12)
plt.xlabel(label[2], fontsize=12)
plt.ylabel('SPECIES', fontsize=12)
plt.show()
```

### Jointplot for data

```python
print('Jointplot for IRIS Data')
sns.jointplot(data1,data1,s=100, alpha=0.5)
plt.show()
```

```python
print('Jointplot for {} vs {}'.format(label[0],label[1]))
sns.jointplot(data[column[0]], data[column[1]], 
              kind='scatter',
              s=200, alpha=0.5)
plt.show()
```

### Pairplot for data and its features

```python
sns.pairplot(data1)
plt.show()
```

```python
print('Pairplot for {} vs {}'.format(label[0], label[1]))
sns.pairplot(data1.drop([column[2], column[3]], axis=1), kind='reg')
plt.show()
```

```python
print('Pairplot for {} vs {}'.format(label[2], label[3]))
sns.pairplot(data1.drop([column[0], column[1]], axis=1))
plt.show()
```

### Countplot for data and its features

```python
for i in range(len(color)):
    plt.figure(figsize=(10,5))
    plt.title('Countplot for {}'.format(label[i]), fontsize=15)
    sns.countplot(x=column[i], data=data1)
    plt.xlabel(label[i], fontsize=12)
    plt.show()
```

```python
for i in range(len(color)):
    plt.figure(figsize=(10,5))
    plt.title('Countplot for {} by Category'.format(label[i]), fontsize=15)
    sns.countplot(x=column[i], data=data1, hue=data['species'])
    plt.xlabel(label[i], fontsize=12)
    plt.show()
```

```python
for i in range(len(color)):
    plt.figure(figsize=(10,5))
    plt.title('Countplot for {}'.format(label[i]), fontsize=15)
    sns.countplot(x=column[i], data=data1, palette='Set1')
    plt.xlabel(label[1], fontsize=12)
    plt.show()
```

### Pointplot for data and its features

```python
plt.figure(figsize=(10,5))
plt.title('Pointplot for {} vs {} by Category'.format(label[0], label[1]), fontsize=15)
sns.pointplot(x=column[0], y=column[1], data=data, hue='species')
plt.xlabel(label[0], fontsize=12)
plt.ylabel(label[1], fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Pointplot for {} vs {} by Category'.format(label[2], label[3]), fontsize=15)
sns.pointplot(x=column[0], y=column[1], data=data, hue='species')
plt.xlabel(label[2], fontsize=12)
plt.ylabel(label[3], fontsize=12)
plt.show()
```
